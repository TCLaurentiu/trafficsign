```python
import torch
import matplotlib.pyplot as plt
```

YOLOV5 small model, that achieves good performance, but is too slow for the Raspberry PI:

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', 'best.pt')
```

Get all paths in "test - pictures" folder:

```python
import os
images = os.listdir("test - pictures")
```

Run the model on each picture and display bounding boxes:

```python
%matplotlib inline
for image in images:
  output = model('test - pictures/' + image)
  print(output)
  print(output.pandas().xyxy[0])
  image = plt.imread('test - pictures/' + image)
  plt.imshow(image)
  for index, row in output.pandas().xyxy[0].iterrows():
    xmin, ymin, xmax, ymax, confidence, _, name = row
    if confidence < 0.3:
      continue
    rectangle = plt.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin, color = "r", alpha = 0.5, fill = False)
    plt.gca().add_patch(rectangle)
    t = plt.text(xmin, ymin, name, c = "w", fontsize = 6)
    t.set_bbox(dict(facecolor='blue', alpha=0.5, edgecolor='red'))
  plt.show()
```
